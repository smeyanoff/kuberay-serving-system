## Get Data Dtypes

```python
dtypes = df.drop(["date", "D_12", "D_24", "D_36", "D_48", "D"], axis=1).dtypes.to_dict()
```

```python
from typing import Optional
```

```python
for col in dtypes.keys():
    dtypes[col] = Optional[type(dtypes[col])]
```

```python
def iter_dict(dict):
    for key in dict.keys():
        yield f"{key}: {dict[key]} \n"
```

```python
iterator = iter_dict(dtypes)
```

```python
with open("./dtypes", "w") as file:
    file.writelines(iterator)
```
